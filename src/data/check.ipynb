```python
from enum import Enum
import pandas as pd
import math
```

```python
class Decision(Enum):
    HOLD = 0
    BUY = 1
    SELL = 2
```

```python
df = pd.read_csv("RELIANCE.csv")
```

```python
df.head(50)
```

```python
history = []
period = 20
num_std_dev = 2.0
```

```python
def generate_signal(current_price: float):
    """Calculates Bollinger Bands and outputs a trading decision."""
    # We need enough historical data to calculate a valid variance
    if len(history) < period:
        return Decision.HOLD

    # 1. Calculate Simple Moving Average (SMA)
    sma = sum(history) / period

    # 2. Calculate Standard Deviation
    variance_sum = sum((p - sma) ** 2 for p in history)
    std_dev = math.sqrt(variance_sum / period)

    # 3. Calculate Bands
    upper_band = sma + (num_std_dev * std_dev)
    lower_band = sma - (num_std_dev * std_dev)

    print(f"Current Price : {current_price}")
    print(f"Upper Band : {upper_band}")
    print(f"Lower Band : {lower_band}")

    # 4. Generate Mean Reversion Signals
    if current_price >= upper_band:
        # return Decision.SELL
        print("SELL")
        return
    elif current_price <= lower_band:
        # return Decision.BUY
        print("BUY")
        return

    # return Decision.HOLD
    print("HOLD")
```

```python
close_prices = df['Close'].tolist()
open_prices = df['Open'].tolist()
```

```python
# First 20 values
history = close_prices[:20]
```

```python
# print(history)
for i in range(20,50):
    print(f"Index : {i}")
    history = close_prices[i-20:i]
    generate_signal(close_prices[i])
    print('-'*20)
```
